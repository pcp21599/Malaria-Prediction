```python
import pandas as pd
```

```python
data=pd.read_csv("Data.csv")
frame=pd.DataFrame(data)
Total = frame["Total_Case"]
Outbreak=[]
```

```python
for x in Total:
    if x>200:
        Outbreak.append("Yes")
    else:
        Outbreak.append("No")
```

```python
Outbreak = pd.DataFrame(Outbreak)
```

```python
frame["Outbreak"] =Outbreak
```

```python
frame.to_csv("Final Data.csv",index= False)
Outbreak
```
