```python
import pandas as pd
# visualization
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline

```

```python
dataSet = ["Scapped Data\Aligarh 2016.csv","Scapped Data\Aligarh 2017.csv","Scapped Data\Bulandshahar 2016.csv","Scapped Data\Bulandshahar 2017.csv","Scapped Data\Ghaziabad 2016.csv","Scapped Data\Ghaziabad 2017.csv","Scapped Data\Greater_Noida-2016.csv","Scapped Data\Greater_Noida-2017.csv","Scapped Data\Meerut-2016.csv","Scapped Data\Meerut-2017.csv"]
combined_csv = pd.concat( [ pd.read_csv(f) for f in dataSet] )
combined_csv.to_csv( "combined_csv.csv", index=False )
    
```

```python
data=pd.DataFrame(combined_csv)
data = data.drop(columns="Month")
data = data.drop(columns="Unnamed: 0")
data.to_csv("Data.csv",index= False)
```
